```python
def missing (train):
    missing_number = df.isnull().sum().sort_values(ascending=False)
    missing_percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)
    missing_values = pd.concat([missing_number, missing_percent], axis=1, keys=['Missing_Number', 'Missing_Percent'])
    return missing_values
missing(df)
```

```python
df['Embarked'].unique()
```

```python
df.Sex=df.Sex.map({'female':0, 'male':1})
df.Embarked=df.Embarked.map({'S':0, 'C':1, 'Q':2,1:1,2:2,0:0,'nan':'NaN'})
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(value=df['Age'].median(),inplace=True)
df['Embarked'].fillna(value=df['Embarked'].median(),inplace=True)
df.Embarked.isnull().sum()
```

```python
X=df.iloc[:,[2,4,5,6,7,9,11]].values
y=df.iloc[:,1] 
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5,metric='euclidean') 
```

```python
model.fit(X,y)
```

```python
y_p=model.predict(X)
y_p
```

```python
y
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y,y_p)
```

```python
confusion_matrix(y_p,y)
```

```python
print(classification_report(y_p,y))
```

```python
# KMeans Clustering
X=df.iloc[:,[2,4,5,6,7,9]].values
y=df.iloc[:,1].values
# df.isnull().sum()
```
